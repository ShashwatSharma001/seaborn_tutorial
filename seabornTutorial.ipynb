```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python

df=sns.load_dataset("iris")
```

```python
df.head()
```

# correlation with Heatmap.
The property of correlation: 
it will only take features between -1 to +1 and do calculations with int or float.
as we know correlation donot perform calculation in accordance to our catagorical features.

```python
df.corr()
```

## Here we have a heatmap

```python
sns.heatmap(df.corr())
```

# join plot
join plot allows to study the relationship between 2 numerical values.

## Bivariate Analysis

```python
sns.jointplot(x='petal_length', y= 'sepal_length', data=df,kind='hex')
```

```python
# using regression
```

```python
sns.jointplot(x='petal_length', y= 'sepal_length', data=df,kind='reg')
```

## pair plot

```python
sns.pairplot(df)
```

## Dist plot

```python
sns.distplot(df["petal_length"])
```
